# Zadanie 4

```python
from utils import *
import random

xor_two_X = [[1, 1], [1, 0], [0, 1], [0, 0]]
xor_two_Y = [[0], [1], [1], [0]]
learning_rate = 0.01

network = NeuralNetwork(hidden_layer=[Sigmoid(weights=[random.uniform(-1, 1), random.uniform(-1, 1),], bias=random.uniform(-1, 1), learning_rate=learning_rate),
                                      Sigmoid(weights=[random.uniform(-1, 1),random.uniform(-1, 1)], bias=random.uniform(-1, 1), learning_rate=learning_rate)],
                        output_layer=[Sigmoid(weights=[random.uniform(-1, 1),random.uniform(-1, 1)], bias=random.uniform(-1, 1), learning_rate=learning_rate)],
                        learnign_rate=0.25)

network.train(xor_two_X, xor_two_Y, 100_000)

create_plot_with_two(xor_two_X, [0, 1, 1, 0], [network.hidden_layer[0], network.hidden_layer[1]])

print(network.hidden_layer[0].initial_weights, network.hidden_layer[0].initial_bias)
print(network.hidden_layer[1].initial_weights, network.hidden_layer[1].initial_bias)
print(network.output_layer[0].initial_weights, network.output_layer[0].initial_bias)
for x, y in zip(xor_two_X, xor_two_Y):
    print(f"{network.forward_propagation(x)} - {y}")
```

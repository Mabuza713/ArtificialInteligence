# Zadanie 4 - same sigmoidy

```python
from utils import *
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)

y_encoded = one_hot_encode(y)

learning_rate = 0.05
n_inputs = 4 #4 wejscia
n_hidden = 4
n_outputs = 3 #3 wyjscia

network = NeuralNetwork(
    hidden_layer=[Sigmoid(weights=[random.uniform(-1,1) for _ in range(n_inputs)],
                          bias=random.uniform(-1,1), learning_rate=learning_rate)
                  for _ in range(n_hidden)],
    output_layer=[Sigmoid(weights=[random.uniform(-1,1) for _ in range(n_hidden)],
                          bias=random.uniform(-1,1), learning_rate=learning_rate)
                  for _ in range(n_outputs)],
    learnign_rate=learning_rate
)

network.train(X.tolist(), y_encoded.tolist(), 10_000)

acc = []

for i in range(len(X)):
    input_sample = X[i].tolist()
    predicted = network.forward_propagation(input_sample)
    actual_class = y_encoded[i].tolist()
    input_sample_fmt = format_list(input_sample)
    predicted_fmt = format_list(predicted, 3)
    actual_class_fmt = format_list(actual_class)
    acc_check(predicted_fmt, actual_class_fmt, acc)
    print(f"Wejście: {input_sample_fmt} -> Predykcja: {predicted_fmt} -> Rzeczywista: {actual_class_fmt}")

acc_rate = len(acc) / len(X)
print(f'accuracy: {acc_rate*100:.1f}%')
```

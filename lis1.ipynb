1.1

```python
x1 = [0.6]
w1 = [1.0]

x2 = [0.2, 0.6]
w2 = [0.7, 0.3]

b = 0.5
```

```python
def unipolar(x, w, prog):
    suma_wazona = 0
    for i in range(len(x)):
        suma_wazona += x[i] * w[i]
    if suma_wazona > prog:
        return 1
    else:
        return 0
```

```python
def bipolar(x, w, prog):
    suma_wazona = 0
    for i in range(len(x)):
        suma_wazona += x[i] * w[i]
    if suma_wazona > prog:
        return 1
    else:
        return -1
```

```python
unipolar(x1, w1, b)
```

```python
unipolar(x2, w2, b)
```

```python
bipolar(x1, w1, b)
```

```python
bipolar(x2, w2, b)
```

```python
def perceptron(x1, x2, w, y_o, prog, u):
    print('x1', 'x2', 'y', 'w', 'prog', 'blad')
    for epoch in range(1000):
        total_error = 0
        for i in range(len(x1)):
            suma_wazona = x1[i]*w[0] + x2[i]*w[1] + prog
            y = 1 if suma_wazona > 0 else 0
            e = y_o[i] - y
            total_error += abs(e)

            w[0] = w[0]+u*e*x1[i]
            w[1] = w[1]+u*e*x2[i]
            prog = prog+u*e

            print(x1[i], x2[i], y, w, prog, e)

        if total_error == 0:
            print(f'Uczenie zakończone po {epoch+1} epokach')
            break
```

```python
import random
x1_or = [0, 0, 1, 1]
x2_or = [0, 1, 0, 1]
y_oczekiwany = [0, 1, 1, 1]
random.seed(0)
w = [round(random.random(),2), round(random.random(),2)]
b = 0.5
u = 0.1

perceptron(x1_or, x2_or, w, y_oczekiwany, b, u)
```

```python
x1_xor = [0, 0, 1, 1]
x2_xor = [0, 1, 0, 1]
y_oczekiwany = [0, 1, 1, 0]
random.seed(0)
w1 = [round(random.random(),2), round(random.random(),2)]
w2 = [round(random.random(),2), round(random.random(),2)]
b1 = 0.5
b2 = 0.7
u_out = 0.1
u = 0.15
w_out = [round(random.random(), 2), round(random.random(),2)]
b_out = 0.125

import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_pochodna(x):
    s = sigmoid(x)
    return s * (1 - s)

def siec_neuronowa(x1, x2, y_oczekiwany, w1, w2, w_out, b1, b2, b_out, u_out, u):
    result = []
    for epoch in range(10001):
        total_error = 0
        for i in range(len(x1)):
            s1 = x1[i]*w1[0] + x2[i]*w1[1] + b1
            s2 = x1[i]*w2[0] + x2[i]*w2[1] + b2
            y1 = sigmoid(s1)
            y2 = sigmoid(s2)
            s_out = y1*w_out[0] + y2*w_out[1] + b_out
            y_out = sigmoid(s_out)
            result.append(y_out)

            e = y_oczekiwany[i] - y_out
            delta_out = e * sigmoid_pochodna(s_out)
            delta1 = delta_out * w_out[0] * sigmoid_pochodna(s1)
            delta2 = delta_out * w_out[1] * sigmoid_pochodna(s2)

            w_out[0] += u_out * delta_out * y1
            w_out[1] += u_out * delta_out * y2
            b_out += u_out * delta_out

            w1[0] += u * delta1 * x1[i]
            w1[1] += u * delta1 * x2[i]
            b1 += u * delta1

            w2[0] += u * delta2 * x1[i]
            w2[1] += u * delta2 * x2[i]
            b2 += u * delta2

            total_error += abs(e)

        print(f'Epoka {epoch}, błąd: {round(total_error,4)}')

    print('Końcowe wyniki:')
    for i in range(len(x1)):
        s1 = x1[i]*w1[0] + x2[i]*w1[1] + b1
        s2 = x1[i]*w2[0] + x2[i]*w2[1] + b2
        y1 = sigmoid(s1)
        y2 = sigmoid(s2)
        s_out = y1*w_out[0] + y2*w_out[1] + b_out
        y_out = sigmoid(s_out)
        print(f'Wejście: ({x1[i]}, {x2[i]}) → Wyjście: {round(y_out, 3)}')

siec_neuronowa(x1_xor, x2_xor, y_oczekiwany, w1, w2, w_out, b1, b2, b_out, u_out, u)
```
